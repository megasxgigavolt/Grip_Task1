```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data.head()
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
array = data.values
X = array[:,0]
Y = array[:,-1]
t_size = 0.2
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=t_size)
```

```python
plt.scatter(X_train, Y_train, color = "red")
plt.title("Relationship between Student's work effort to his/her performance")
plt.xlabel("Number Of Hours")
plt.ylabel("Score")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(X_train.reshape(-1,1),Y_train)
pred = lg.predict(X_test.reshape(-1,1))
```

```python
plt.scatter(X_train, Y_train, color = "red")
plt.plot(X_train, lg.predict(X_train.reshape(-1,1)), color = "green")
plt.title("Relationship between Student's work effort to his/her performance")
plt.xlabel("Number Of Hours")
plt.ylabel("Score")
plt.show()
```

```python
#from sklearn.metrics import accuracy_score
lg.score(X_test.reshape(-1,1),Y_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("mean absolute error = ",mean_absolute_error(Y_test,pred))
```

```python
hours = 9.25
own_pred = lg.predict(np.array(hours).reshape(1,-1))
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
